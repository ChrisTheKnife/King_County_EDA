```python
import warnings

warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Loading the dataset:
basic_king = pd.read_csv('data/King_County_House_prices_dataset.csv')

# Get info about the variables
basic_king.info()
```

```python
df2 = pd.DataFrame({
    'Factor': ['Growth', 'Value'],
    'Weight': [0.10, 0.20],
    'Variance': [0.15, 0.35]
})
fig, ax1 = plt.subplots(figsize=(10, 10))
tidy = df2.melt(id_vars='Factor').rename(columns=str.title)
sns.barplot(x='Factor', y='Value', hue='Variable', data=tidy, ax=ax1)
sns.despine(fig)
```

```python
appl_price = [110.1, 122.5, 150, 168, 184.8]
ms_price = [90.1, 132.5, 150, 125.7, 170.8]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year, appl_price, 'k',
        year, ms_price, 'g')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.show()
```

```python
plt.plot(year, appl_price, '--k',
        year, ms_price, 'og')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.show()
```

```python
from matplotlib.ticker import MaxNLocator

fig_1 = plt.figure(1, figsize=(20, 4.8))

chart_1 = fig_1.add_subplot(121)
chart_2 = fig_1.add_subplot(122)

chart_2.xaxis.set

chart_1.plot(year, appl_price)
chart_1.xaxis.set_major_locator(MaxNLocator(integer=True))
chart_2.scatter(year, ms_price)
chart_2.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()
```

```python
fig_2, axes = plt.subplots(2,2, figsize=(20,5))
axes[0,1].scatter(year, ms_price)
axes[0,1].xaxis.set_major_locator(MaxNLocator(integer=True))
axes[1,0].plot(year, appl_price)
axes[1,0].xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()
```
